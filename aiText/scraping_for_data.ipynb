## Heste-nettet nyheder

```python
def extract_text_HNN(link, headers):

    response = requests.get(link, headers=headers)
    # contains danish characters
    response.encoding = "utf-8"

    soup = BeautifulSoup(response.content, "html.parser")

    # print(soup.prettify())

    # get main text
    main_text = soup.find("div", {"class": "content2_2"})

    # get title
    title = main_text.find("h1").text

    # get articleBody
    articleBody = main_text.find("div", {"itemprop": "articleBody"}).text
    # the relevant text looks like: ></div><p><meta content="2023-03-14T12:07:42" itemprop="datePublished"><div>Tilf√∏jet d. 14-03-2023 kl. 12:07 af <span itempr
    date = soup.find("meta", {"itemprop": "datePublished"})["content"]

    # combine title and articleBody
    article = title + '\n' + date + '\n' + articleBody

    return article

link_base = "https://www.heste-nettet.dk/nyheder/"
out_path = 'data/heste-nettet-nyheder/'

article_nums = np.arange(15282-8000, 15282)
for num in article_nums:
    fname = out_path + str(num) + '.txt'
    try:
        # check if exists
        if not os.path.isfile(fname):
            
            article = extract_text_HNN(link_base+str(num), headers)

            # save article
            
            with open(fname, 'w', encoding='utf-8') as f:
                f.write(article)

    except AttributeError:
        print('Article number {} not found'.format(num))
```

## Heste-nettet forum

```python

def extract_text_HNF(link, headers):
    response = requests.get(link, headers=headers)
    # contains danish characters
    response.encoding = "utf-8"

    soup = BeautifulSoup(response.content, "html.parser")
    try:
        title = soup.find('title').text
    except AttributeError:
        title = 'No title found'


    # the relevant parts are here:
    # <div class=\"rsPost
    
    main_text = soup.find("div", {"class": "rsPost"}).find("body").text
    main_text
    try:
        date = soup.find("meta", {"itemprop": "datePublished"})["content"]
    except AttributeError:
        date = 'No date found'

    # combine title and articleBody
    article = title + '\n' + date + '\n' + main_text

    return article

link_base = "https://www.heste-nettet.dk/forum/1/"

out_path = 'data/heste-nettet-forum/'
found = 0
article_nums = np.arange(0, 2431998)
for num in article_nums:
    fname = out_path + str(num) + '.txt'
    try:
        # check if exists
        if not os.path.isfile(fname):
            
            article = extract_text_HNF(link_base+str(num)+"/"+str(num), alt_headers)

            # save article
            
            with open(fname, 'w', encoding='utf-8') as f:
                f.write(article)
                found += 1

    except AttributeError as e:
        print(e)
        print('Article number {} not found'.format(num))

    
    
    if found % 1000 == 0:
        print('Found {} articles'.format(found))
        break



    
```
